```python
import pandas as pd

df = pd.read_csv('../vehicles_us.csv', sep=',')

print(df.head())
```

```python
df.info()
```

```python
# 1. Calcule a média e arredonde para o ano inteiro mais próximo
media_ano = df['model_year'].mean().round(0)

# 2. Preencha os valores ausentes (NaN) com essa média
df['model_year'] = df['model_year'].fillna(media_ano)

# 3. Verifique se os nulos desapareceram na coluna 'model_year'
print(df['model_year'].isna().sum())
```

```python
# Converte a coluna 'model_year' para inteiro
df['model_year'] = df['model_year'].astype('int64')

# Exibe o info() novamente para conferir o resultado
df.info()
```

```python
# Média da quilometragem
media_odo = df['odometer'].mean().round(0)

# Mediana da quilometragem (valor central)
mediana_odo = df['odometer'].median().round(0)

print(f"Média do Odômetro: {media_odo:.0f}")
print(f"Mediana do Odômetro: {mediana_odo:.0f}")
```

```python
# 1. Calcule a mediana e arredonde para o inteiro mais próximo
mediana_odo = df['odometer'].median().round(0)

# 2. Preencha os valores ausentes (NaN) com essa mediana
df['odometer'] = df['odometer'].fillna(mediana_odo)

# 3. Verifique se os nulos desapareceram na coluna 'odometer'
print(df['odometer'].isna().sum())
```

```python
# Converte a coluna 'odometer' para inteiro
df['odometer'] = df['odometer'].astype('int64')

df.info()
```

```python
moda_cylinders = df['cylinders'].mode()[0]

print(f"Moda (Cilindro mais comum): {moda_cylinders:.0f}")
```

```python
df['cylinders'] = df['cylinders'].fillna(8)

# 2. Converte a coluna 'cylinders' para inteiro (int64)
df['cylinders'] = df['cylinders'].astype('int64')

df.info()
```

```python
# 1. Preenche os valores nulos (NaN) com 0, assumindo que a ausência significa que não é 4x4
df['is_4wd'] = df['is_4wd'].fillna(0)

# 2. Converte a coluna 'is_4wd' para o tipo inteiro (int64)
df['is_4wd'] = df['is_4wd'].astype('int64')

df.info()
```

```python
df['paint_color'] = df['paint_color'].fillna('desconhecida')

print(df['paint_color'].isna().sum())
```

```python
df['date_posted'] = pd.to_datetime(df['date_posted'])

df.info()
```

```python
print(df['price'].describe())
```

```python
preco_min = 500
preco_max = 50000

df_filtrado = df.query('@preco_min <= price <= @preco_max')

linhas_removidas = len(df) - len(df_filtrado)
print(f"Total de linhas originais: {len(df)}")
print(f"Linhas removidas (outliers de preço): {linhas_removidas}")

# Verifique as novas estatísticas de preço no DataFrame filtrado
print("\nNovas Estatísticas de Preço:")
print(df_filtrado['price'].describe())
```

```python
print(df_filtrado['odometer'].describe())
```

```python
# 1. Definindo os novos limites
odo_min = 1000
odo_max = 350000

df_filtrado = df_filtrado.query('@odo_min <= odometer <= @odo_max')

# 3. Imprime o número de linhas removidas desta vez
linhas_removidas_odo = 50229 - len(df_filtrado) 
print(f"Linhas removidas (outliers de odômetro): {linhas_removidas_odo}")

# 4. Verifique as novas estatísticas de odômetro
print("\nNovas Estatísticas de Odômetro:")
print(df_filtrado['odometer'].describe())
```

```python
# Agrupa o DataFrame por 'condition' e calcula o preço médio
preco_medio_por_condicao = df_filtrado.groupby('condition')['price'].mean().sort_values(ascending=False)

print("Preço Médio por Condição:")
print(preco_medio_por_condicao)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração do gráfico
plt.figure(figsize=(10, 6))

# Cria o gráfico de barras
sns.barplot(x=preco_medio_por_condicao.index, y=preco_medio_por_condicao.values, palette="viridis")

# Adiciona títulos e rótulos
plt.title('Preço Médio dos Veículos por Condição', fontsize=16)
plt.xlabel('Condição', fontsize=12)
plt.ylabel('Preço Médio (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
ANO_ATUAL = 2020

# Cria a coluna 'age' (idade do veículo)
df_filtrado['age'] = ANO_ATUAL - df_filtrado['model_year']

print("Estatísticas da Idade do Veículo:")
print(df_filtrado['age'].describe())
```

```python
df_filtrado['age_for_calc'] = df_filtrado['age'].apply(lambda x: 1 if x == 0 else x)

# coluna 'km_per_year'
df_filtrado['km_per_year'] = df_filtrado['odometer'] / df_filtrado['age_for_calc']

# Arredonda a coluna para facilitar a leitura
df_filtrado['km_per_year'] = df_filtrado['km_per_year'].round(0).astype('int64')

print("\nEstatísticas da Quilometragem por Ano:")
print(df_filtrado['km_per_year'].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração da figura
plt.figure(figsize=(14, 8))

# Cria o gráfico de caixa, usando a paleta 'tab10'
sns.boxplot(
    data=df_filtrado, 
    x='type', 
    y='price', 
    palette='tab10' # Paleta alterada para 10 cores contrastantes
)

# Adiciona títulos e rótulos
plt.title('Distribuição de Preços por Tipo de Veículo', fontsize=16)
plt.xlabel('Tipo de Veículo', fontsize=12)
plt.ylabel('Preço (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 1. Cria a coluna 'manufacturer' (fabricante)
df_filtrado['manufacturer'] = df_filtrado['model'].apply(lambda x: x.split(' ')[0])

# 2. Verifica quantos fabricantes únicos existem
print(f"Total de Fabricantes Únicos: {df_filtrado['manufacturer'].nunique()}")

# 3. Cria a tabela de resumo (Pivot Table):
# Conta o número de veículos por Fabricante e por Tipo
contagem_fabricante_tipo = df_filtrado.pivot_table(
    index='manufacturer', 
    columns='type', 
    values='price', 
    aggfunc='count',
    fill_value=0
)

print("\nTabela de Contagem por Fabricante e Tipo (Top 5 Fabricantes):")
print(contagem_fabricante_tipo.head())
```

```python
# 1. Renomeia
df_final = df_filtrado

# 2. Salva o DataFrame final em um novo arquivo CSV, sem o índice do Pandas
df_final.to_csv('vehicles_final.csv', index=False)

print(f"O DataFrame final com {len(df_final)} linhas foi salvo em 'vehicles_final.csv'.")
print("O arquivo está pronto para ser usado no seu aplicativo web!")
```
